### Remark 2.31 (Gaussians).

Suppose $\alpha \sim \mathcal{N}( m_\alpha, \Sigma_\alpha)$ and $\beta \sim \mathcal{N}(m_\beta, \Sigma_\beta)$.

Then, one can show that the following map:

$$
T : x \mapsto m_\beta + A( x - m_\alpha) ,
$$

where

$$
A = \Sigma_\alpha^{-1/2} ( \Sigma_\alpha^{1/2} \Sigma_\beta \Sigma_\alpha^{1/2} ) ^{1/2} \Sigma_\alpha^{-1/2}  \equiv A^T
$$
